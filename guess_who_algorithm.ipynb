```python
def calc_entropy(column):
    """
    Calculate entropy.
    :param column: Pandas Series, list, or Numpy array
    :return:
    """
    counts = column.value_counts()
    probabilities = counts/len(column)
    entropy = 0
    for p in probabilities:
        if p > 0:
            entropy += p * math.log(p, 2)
    return -entropy

def calc_information_gain(data, split_name, target_column):
    """
    Calculate information gain for dataset with binary values.
    :param data: Pandas DataFrame with binary values
    :param split_name: name of column to split on
    :param target_column: target column
    :return:
    """
    orig_entropy = calc_entropy(data[target_column])
    col = data[split_name]
    left_split = data[col == 0]
    right_split = data[col == 1]
    to_subtract = 0
    for subset in [left_split, right_split]:
        probability = subset.shape[0]/data.shape[0]
        to_subtract += probability * calc_entropy(subset[target_column])
    return orig_entropy - to_subtract

def find_best_column(data, target_column):
    """
    Find the best column to split on.
    :param data: Pandas DataFrame with column names
    :param target_column: name of target column
    :return: name of best column
    """
    # list columns to potentially split on
    columns = data.columns.values.tolist()
    columns.remove(target_column)

    #find column with the highest information gain
    info_gain_dict = {}
    for col in columns:
        info_gain_dict[col] = calc_information_gain(data, col, target_column)
    highest_gain = max(info_gain_dict, key=info_gain_dict.get)
    return highest_gain

tree = {}
nodes = []

def tree_builder(data, target_column, tree):
    """
    Recursively builds a binary decision tree. Stores it as a dictionary and outputs it to a .txt file.
    :param data:
    :param target_column:
    :param tree:
    :return:
    """
    columns = data.columns.values.tolist()
    columns.remove(target_column)

    unique_targets = pd.unique(data[target_column])

    nodes.append(len(nodes) + 1)
    # tree['node_number'] = nodes[-1]

    if len(unique_targets) == 1:
        tree['SOLUTION'] = list(data['name'].values)
        return

    best_column = find_best_column(data, target_column)
    tree['next_question'] = best_column
    # tree['remaining names'] = list(data['name'].values)


    left_split = data[data[best_column] == 0]
    right_split = data[data[best_column] == 1]
    split_dict = [["NO", left_split], ["YES", right_split]]

    for name, split in split_dict:
        tree[name] = {}
        tree_builder(split, target_column, tree[name])

    return tree

guess_who_dict = tree_builder(guess_who_df, 'name', tree)

with open('guess_who_decision_tree.txt', 'w') as convert_file:
     convert_file.write(json.dumps(guess_who_dict))
```
